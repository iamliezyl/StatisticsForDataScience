```python
import pandas as pd
import seaborn as sns
import numpy as  np
import matplotlib.pyplot as plt
```

# USING SCATTERPLOT

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
tips.describe()
```

```python
# Using ScatterPlot
# To see if there's a relationship between two variables

sns.scatterplot(x='total_bill', y='tip', data = tips, hue='sex')
```

```python
# in this scatterplot, it shows that theres a linear relationship.  As the bill increases, the tip also increases
# adding hue for sex shows a multivariate relationship analysis
```

```python
# lmplot shows and overall linear relationship 

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
# data points that are far away from the line are considered as outliers
# the convergence point is the statistical mean or the general prediction of the tip value 
#insights: female are more likely to give the usual amount.  There are few male that gives exceptional tips
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'x'])
```

```python
#show on separate facets
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

# USING HEATMAP

```python
# first get the value of correlation

corr = df.corr()
corr
```

```python
sns.heatmap(corr, annot = True)
plt.figure(figsize = (10,10))   # to increase the display size
```

```python
# this figure shows the high positive and negative correlations
```

```python
# to show a different color coding use cmap property
sns.heatmap(corr, annot = True, cmap ='BrBG')
```
